First, let's get a pre-trained model for neurons

```python
from staticnet_analyses import multi_mei
from staticnet_experiments import models

key = {'data_hash': '7572eed73113c993e7d1b92f83e270b4', 'group_id': 29, 
       'net_hash': '80d0d4bc112470b2ba04cd5eba048e39', 'neuron_id': 119, 
       'readout_key': 'group029-21067-9-17-0'}

# Get our f (average of four models)
train_stats = multi_mei.prepare_data(key, key['readout_key'])
_, (_, _, height, width), _, mean_behavior, mean_eyepos, _ = train_stats
model_key = {'group_id': key['group_id'], 'net_hash': key['net_hash']}
my_models = [(models.Model & mk).load_network() for mk in (models.Model & model_key).proj()]
model = featurevis.Ensemble(my_models, key['readout_key'], eye_pos=mean_eyepos, 
                             neuron_idx=key['neuron_id'], device=device)
```

## DiCarlo (Bashivan et al., 2018)
See Sec. *Synthesized "controller" images* in p.9.
* Optimizer: SGD
* Transform: Jittering
* Regularization: Total variation
* Gradient function: Normalize the gradient (grad / norm(grad)) and clip between -1 and 1.

```python
dc_transform = featurevis.Jitter(max_jitter=(2, 4))
dc_regularization = featurevis.TotalVariation(weight=0.001)
dc_gradient = featurevis.Compose([featurevis.ChangeNorm(1), 
                                  featurevis.ClipRange(-1, 1)])
```

```python
opt_x, fevals, reg_values = featurevis.gradient_ascent(model, initial_image, step_size=1, num_iterations=700, 
                                                       transform=dc_transform, regularization=dc_regularization, 
                                                       gradient_f=dc_gradient)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(7*3, 4))
axes[0].plot(fevals)
axes[1].plot(reg_values)
axes[2].imshow(opt_x.squeeze().detach().cpu().numpy())
```

## DeepTune (Abbasi-Asl, 2018)
See Equation in p.8
* Optimizer: SGD
* Regularization: total variation and l6 norm

```python
dt_regularization = featurevis.Combine([featurevis.TotalVariation(weight=0.001), 
                                        featurevis.LpNorm(weight=1, p=6)])
```

```python
opt_x, fevals, reg_values = featurevis.gradient_ascent(model, initial_image, step_size=1, num_iterations=500, 
                                                       regularization=dt_regularization)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(7*3, 4))
axes[0].plot(fevals)
axes[1].plot(reg_values)
axes[2].imshow(opt_x.squeeze().detach().cpu().numpy())
```

## Ours (Walker et al., 2019)
* Optimizer: SGD
* Gradient: Fourier smoothing, divide by absolute mean of gradient, multiply by (step_gain/255) and multiply by a decaying learning rate
* Post update: Clip range and blur image with a decaying sigma

```python
walker_gradient = featurevis.Compose([featurevis.FourierSmoothing(0.04), # not exactly the same as fft_smooth(precond=0.1) but close
                                       featurevis.DeepDrawUpdateFactor(0.1),
                                       featurevis.MultiplyBy(3, decay_factor=(3 - 0.125) /(1-1000))])  # decays from 3 to 0.125 in 1000 iterations 
bias, scale = 111.28329467773438, 60.922306060791016
walker_postup = featurevis.Compose([featurevis.ClipRange(-bias / scale, (255 - bias) / scale), 
                                     featurevis.GaussianBlur(1.5, decay_factor=(1.5 - 0.01) /(1-1000))]) # decays from 1.5 to 0.01 in 1000 iterations 
```

```python
opt_x, fevals, reg_values = featurevis.gradient_ascent(model, initial_image, step_size=1, num_iterations=1000, 
                                                       post_update=walker_postup, gradient_f=walker_gradient)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
axes[0].plot(fevals)
axes[1].imshow(opt_x.squeeze().detach().cpu().numpy())
```

## Most exciting Gabor
* Optimizer: Adam (worked better than SGD in this example)
* Input: orientation, phase, wavelength, sigma, dx, dy
* Transform: change parameters range to a nice range and then generate the Gabor with them.

```python
initial_x = torch.randn(1, 6, dtype=torch.float32, device=device) # orientation, phase, wavelength, sigma, dx, dy
```

```python
import math
min_x = torch.tensor([0, 0, 0.12, 0.04, -0.5, -0.5], dtype=torch.float32, device=device)
max_x = torch.tensor([math.pi, 2 * math.pi, 1.5, 1, 0.5, 0.5], dtype=torch.float32, device=device)
gabor_transform = featurevis.Compose([featurevis.ChangeRange(min_x, max_x), featurevis.GaborGenerator(36, 62)])
```

```python
opt_x, fevals, reg_values = featurevis.gradient_ascent(model, initial_x, optim_name='Adam', step_size=1, 
                                                       num_iterations=200, transform=gabor_transform)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 5))
axes[0].plot(fevals)
axes[1].imshow(gabor_transform(opt_x).squeeze().detach().cpu().numpy())
```
